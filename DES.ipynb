```python
import os

# Функція для генерації 8 байтів випадкових даних (64 біти) для IV
def generate_iv():
    return os.urandom(8)

# Спрощена версія DES (демонстраційна, не повна)
def des_encrypt_block(block, key):
    return bytes([b ^ k for b, k in zip(block, key)])

def des_decrypt_block(block, key):
    return bytes([b ^ k for b, k in zip(block, key)])  # Для спрощеного DES

# Функція XOR для двох блоків байтів
def xor_blocks(block1, block2):
    return bytes([b1 ^ b2 for b1, b2 in zip(block1, block2)])

# Реалізація DES шифрування в режимі CBC
def des_cbc_encrypt(plaintext, key, initializationVector):
    if len(plaintext) % 8 != 0:
        raise ValueError("Довжина plaintext повинна бути кратною 8 байтам")

    cipherText = b""
    previousBlock = initializationVector

    for i in range(0, len(plaintext), 8):
        block = plaintext[i:i+8]
        blockToEncrypt = xor_blocks(block, previousBlock)
        encryptedBlock = des_encrypt_block(blockToEncrypt, key)
        cipherText += encryptedBlock
        previousBlock = encryptedBlock

    return cipherText

# Реалізація DES розшифрування в режимі CBC
def des_cbc_decrypt(ciphertext, key, initializationVector):
    if len(ciphertext) % 8 != 0:
        raise ValueError("Довжина ciphertext повинна бути кратною 8 байтам")

    plaintext = b""
    previousBlock = initializationVector

    for i in range(0, len(ciphertext), 8):
        block = ciphertext[i:i+8]
        decryptedBlock = des_decrypt_block(block, key)
        originalBlock = xor_blocks(decryptedBlock, previousBlock)  # XOR з попереднім блоком
        plaintext += originalBlock
        previousBlock = block

    return plaintext

# Генеруємо ключ (8 байтів для DES)
key = os.urandom(8)
# Генеруємо ініціалізаційний вектор (IV)
initializationVector = generate_iv()

# Тестовий текст для шифрування (має бути кратний 8 байтам)
plaintext = b"Hello123Hello123"  # 16 байтів, кратно 8

# Шифруємо текст
cipherText = des_cbc_encrypt(plaintext, key, initializationVector)

# Розшифруємо текст
decryptedText = des_cbc_decrypt(cipherText, key, initializationVector)

print(f"Ключ: {key.hex()}")
print(f"Ініціалізаційний вектор (IV): {initializationVector.hex()}")
print(f"Зашифрований текст: {cipherText.hex()}")
print(f"Розшифрований текст: {decryptedText.decode()}")
```
